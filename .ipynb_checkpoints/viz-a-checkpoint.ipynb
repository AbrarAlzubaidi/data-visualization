# fish data visualization


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# fish data
weight: in gram

length: in cm

height: in cm

width: diagonal width in cm  

```python
data = pd.read_csv('Fish.csv')
data
```

## relation between height and width of fish (Positive relationship)


for each fish if height increase it means by default the width of the fish increased

as i see from diagram the longest type is Bream fish with height ~= 18.8 cm 

```python
sns.set_theme(style="darkgrid")

sns.lineplot(x="Width", y="Height",
             hue="Species", style="Species",
             data=data)
```

# relation between type of fish and its average length with weight

as i see most big fish from Pike type and most types has avg length between(32-43)cm with weight between(125-400)g

most common fish type that have big range of length and wieght is Perch fish with purple color


```python
data['avg length'] = (data['Length1']+data['Length2']+data['Length3'])/3
sns.set_theme(style="white")
g = sns.JointGrid(data=data, x="Weight", y="avg length", space=0, ratio=17, hue="Species")
g.plot_joint(sns.scatterplot, size=data["Species"], sizes=(50, 120),
             color="c", alpha=.8, legend=True )
g.plot_marginals(sns.rugplot, height=5, color="c", alpha=.8)
```
